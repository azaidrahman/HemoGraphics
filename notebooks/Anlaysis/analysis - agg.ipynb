```python
import sys
sys.path.append('../../scripts/')
import pandas as pd
from agg_analysis import donation_trends_malaysia
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
# Apply a rolling mean to smooth the data (you can adjust the window size as needed)
donation_trends_malaysia['rolling_mean'] = donation_trends_malaysia['daily'].rolling(window=30).mean()

# Initialize the seaborn style
sns.set(style="darkgrid")

# Increase the figure size for better visibility
plt.figure(figsize=(20, 10))

# Create a line plot of the rolling mean of daily donations
sns.lineplot(x='date', y='rolling_mean', data=donation_trends_malaysia, label='30-Day Rolling Mean of Daily Donations', color='tab:blue')

# Set the date format for the x-axis labels
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Show one tick per year

# Rotate the date labels for better readability
plt.xticks(rotation=45)

# Set labels and title
plt.title('Smoothed Blood Donation Trends in Malaysia', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Donations', fontsize=14)

# Tight layout to adjust for the rotated labels
plt.tight_layout()

# Show the plot
plt.show()
```
