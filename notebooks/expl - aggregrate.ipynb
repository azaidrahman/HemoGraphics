```python
import pandas as pd
import os

raw_data_path = r"../data/raw"
raw_data_files = os.listdir(raw_data_path)

csv_files = [file for file in raw_data_files if file.endswith('.csv')]

dataframes = {}
for csv_file in csv_files:
    file_path = os.path.join(raw_data_path, csv_file)
    dataframe_name = csv_file.replace('.csv', '')
    dataframes[dataframe_name] = pd.read_csv(file_path)

def perform_eda(df):
    
    print("First few rows:")
    print(df.iloc[:,:5].head())
    print("\nInfo:")
    df.info()
    print("\nDescribe:")
    print(df.describe())
    print("\nNull values:")
    print(df.isnull().sum())
    print("\nNumber of duplicate rows:")
    print(df.duplicated().sum())

# Perform EDA on each dataframe in the dictionary
for name, df in dataframes.items():
    print(f"EDA for {name}:")
    perform_eda(df)
    print("\n" + "="*50 + "\n")

```
